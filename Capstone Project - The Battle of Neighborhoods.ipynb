```python
# Get the top 100 venues that are within a radius of 2000 meters.
radius = 2000
LIMIT = 200

venues = []

for lat, long, Dist in zip(df_new['Latitude'], df_new['Longitude'], df_new['Dist']):
    
    # create the API request URL
    url = "https://api.foursquare.com/v2/venues/explore?client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}".format(
        CLIENT_ID,
        CLIENT_SECRET,
        VERSION,
        lat,
        long,
        radius, 
        LIMIT)
    
    # make the GET request
    results = requests.get(url).json()["response"]['groups'][0]['items']
    
    # return only relevant information for each nearby venue
    for venue in results:
        venues.append((
            Dist,
            lat, 
            long, 
            venue['venue']['name'], 
            venue['venue']['location']['lat'], 
            venue['venue']['location']['lng'],  
            venue['venue']['categories'][0]['name']))
```

```python
# convert the venues list into a new DataFrame
venues_df = pd.DataFrame(venues)

# define the column names
venues_df.columns = ['Dist', 'Latitude', 'Longitude', 'VenueName', 'VenueLatitude', 'VenueLongitude', 'VenueCategory']

print(venues_df.shape)
venues_df.head()
```

```python
# check if the results contain "Shopping Mall"
"Dist" in venues_df['VenueCategory'].unique()

# one hot encoding
hn_onehot = pd.get_dummies(venues_df[['VenueCategory']], prefix="", prefix_sep="")

# add Dist column back to dataframe
hn_onehot['Dist'] = venues_df['Dist'] 

# move Dist column to the first column
fixed_columns = [hn_onehot.columns[-1]] + list(hn_onehot.columns[:-1])
hn_onehot = hn_onehot[fixed_columns]

print(hn_onehot.shape)
hn_onehot.head()
```

```python
hn_grouped = hn_onehot.groupby(["Dist"]).mean().reset_index()

print(hn_grouped.shape)
hn_grouped
```

```python
len(hn_grouped[hn_grouped["Café"] > 0])
```

```python
hn_mall = hn_grouped[["Dist","Café"]]
hn_mall.head()
```

```python
# set number of clusters
kclusters = 3

hn_clustering = hn_mall.drop(["Dist"], 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(hn_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10]
```

```python
# create a new dataframe that includes the cluster as well as the top 10 venues for each neighborhood.
hn_merged = hn_mall.copy()

# add clustering labels
hn_merged["Cluster Labels"] = kmeans.labels_
```

```python
hn_merged.head()
```

```python
# merge toronto_grouped with toronto_data to add latitude/longitude for each neighborhood
hn_merged_SM = hn_merged.join(df2.set_index("Dist"), on="Dist")

print(hn_merged_SM.shape)
hn_merged_SM.head() # check the last columns!
```

```python
print(hn_merged_SM.shape)
hn_merged_SM.sort_values(["Cluster Labels"], inplace=True)
hn_merged_SM
```

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i+x+(i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(hn_merged_SM['Latitude'], hn_merged_SM['Longitude'], hn_merged_SM['Dist'], hn_merged_SM['Cluster Labels']):
    label = folium.Popup(str(poi) + ' - Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
