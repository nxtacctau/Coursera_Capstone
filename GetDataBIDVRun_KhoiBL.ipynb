```python
def getlistings(StrID):
    # datetime object containing current date and time
    now = datetime.now()
    dt_string = now.strftime("%d/%m/%Y %H:%M:%S")

    url = 'https://run.bidv.com.vn/home/user-information/' + StrID
       
    source = urllib.request.urlopen(url).read()
    
    soup = BeautifulSoup(source,'lxml')

    try:
        ActiveName = soup.find('p', class_ = 'profile-name title-3')
        print(ActiveName)
        if ActiveName is not None:
            ActiveName = ActiveName.text
        else: 
            ActiveName ="Acct Name Bị lỗi"
    except AssertionError as error:
        ActiveName =""
        print("ActiveName: " + error) 
        
    try:
        Ebib = soup.find('div', class_='user-id').text
        if Ebib is not None:
            Ebib = Ebib[7:]
        else:
            Ebib =""
    except AssertionError as error:
        Ebib =""
        print("Ebib: " + error) 
        
    try: 
        SoKM = float(soup.find('p', class_='stats-font').text.strip())
    except AssertionError as error:
        SoKM =0 
        print("SoKM: " + error) 
    try:
        SoTien = float(soup.find('p', class_='stats-font number-bn').text.strip().replace(",",""))
    except AssertionError as error:
        SoTien =0 
        print("SoTien: " + error) 
    try:
        IdRunner = StrID
    except AssertionError as error:
        IdRunner ="" 
        print("IdRunner: " + error) 

    for tag in soup.find_all("li"):
        if(tag.text.find('Thẻ vàng') != -1):        
            try:
                theVang = tag.text[:tag.text.find('Thẻ vàng')]              
            except AssertionError as error:
                theVang = ""
                print("The vang: " + error) 
        if (tag.text.find('Thẻ đỏ') != -1):
            try:
                theDo = tag.text[:tag.text.find('Thẻ đỏ')]
            except AssertionError as error:
                theDo = ""
                print("The do: " + error) 
        if(tag.text.find('Hạng cá nhân') != -1):
            try:
                HangCaNhan = tag.text[:tag.text.find('Hạng cá nhân')]       
            except AssertionError as error:
                HangCaNhan =""
                print("Hạng cá nhân: " + error) 
        if (tag.text.find('Hạng trong đội') != -1):
            try:
                HangTrongDoi = tag.text[:tag.text.find('Hạng trong đội')]
            except AssertionError as error:
                HangTrongDoi =""
                print("Hạng trong đội: " + error) 
        
    listings = []
    listings.append([ActiveName, Ebib, IdRunner, SoKM, SoTien, theVang, theDo,
                     HangCaNhan, HangTrongDoi,dt_string  ])

   ### print(listings)
    return listings
```

```python
dfbanle = pd.DataFrame() 
###df = pd.DataFrame(columns=["ActiveName","EBIB","ID", "SoKM", "SoTien","Thevang","TheDjo","HangCanhan","HangTrongdoi","Thoidiemtai"] )
dfbanle
for SoID in dataBanle["ID"]:
    if numpy.isnan(SoID):
        print("Số ID is None")        
    else:
        print(str(int(SoID)))
        dfbanle = dfbanle.append(getlistings(str(int(SoID))))
```
